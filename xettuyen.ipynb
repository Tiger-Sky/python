```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Viết hàm nạp dữ liệu và in ra 10 dòng dữ liệu đầu tiên

```python
df= pd.read_csv('dulieuxettuyendaihoc (1).csv')
df.head(10)
```

Kiểm tra các kiểu dữ liệu của các trường

```python
df.dtypes
```

Kiểm tra tổng số dữ liệu trống của các trường dữ liệu

```python
df.isnull().sum()
```

Điền vào dữ liệu trống của cột DT bằng giá trị K

```python
df['DT'].fillna(value='K', inplace=True)
df['DT']
```

chuyển sang dữ liệu chuỗi

```python
df['DT'] = df['DT'].astype("string")
df['DT'].dtype
```

 Chuyển dữ liệu giới tính thành 0 và 1 tương ứng với giới tính Nữ và giới 
tính Nam

```python
df['GT'] = df['GT'].map({'M': 1, 'F': 0})
df['GT']
```

 Chuyển dữ liệu khối thi tương ứng: 'A':0, 'A1':1, 'B':2, 'C':3, 'D1':4

```python
mapping = {'A':0, 'A1':1, 'B':2, 'C':3, 'D1':4}
df['KT'] = df['KT'].map(mapping)
df['KT'].head(20)
```

 Tạo cột Điểm ĐH và KQ  điểm sàn với các thông tin như sau: KQ điểm 
sàn Điểm ĐH=(DH1 + DH2 +DH3); KQ: Nếu Điểm ĐH >= 15 thì  Đạt 
sàn và Điểm ĐH < 15  Không đạt sàn

```python
df['Điểm ĐH'] = df['DH1'] + df['DH2'] + df['DH3']
df['KQ điểm sàn'] = df['Điểm ĐH'].apply(lambda x: 'Đạt sàn' if x >= 15 else 'Không đạt sàn')
df
```

Thống kê các giá trị count, sum, mean, median, min, max, std của 
DH1 theo KT

```python
df.groupby('KT')['DH1'].agg(['count', 'sum', 'mean', 'median', 'min', 'max', 'std'])
```

Thống kê các giá trị count, sum, mean, median, min, max, std của 
DH1 theo KT và KV

```python
df.groupby(['KT', 'KV'])['DH1'].agg(['count', 'sum', 'mean', 'median', 'min', 'max', 'std'])
```

Trình bày dữ liệu với các biến DH1, DH2, DH3 lớn hơn bằng 5 và 
thuộc khu vực 2NT 

```python
result = df[(df['DH1'] >= 5) & (df['DH2'] >= 5) & (df['DH3'] >= 5) & (df['KV'] == '2NT')]

# In kết quả
print(result)
```

Trình bày dữ liệu với các biến DH1, DH2, DH3 lớn hơn bằng 5 và thi 
khối A.

```python
df.loc[(df['DH1'] >= 5) & (df['DH2'] >= 5) & (df['DH3'] >= 5) & (df['KT'] == 'A')]

```

 Tính điểm trung bình 2NT của từng môn và điểm trung bình đại học

```python
mean_2nt = df.loc[:, 'T1':'N6'].mean()


diem_tb_daihoc = df['Điểm ĐH'].mean()


print("Điểm trung bình 2NT của từng môn:\n", mean_2nt)
print("\nĐiểm trung bình đại học:\n", diem_tb_daihoc)
```

Thống kê số lượng học sinh Nam và Nữ theo Dân tộc (DT)

```python
hoc_sinh_theo_dt_va_gt = df.groupby(['DT', 'GT']).size()
print(hoc_sinh_theo_dt_va_gt)
```
